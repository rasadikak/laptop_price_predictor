```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
```

```python
data = pd.read_csv(url, encoding='latin-1')
```

```python
data = data.drop(columns=['laptop_ID'])
```

```python
data.head(2)
```

```python
data['Ram'] = data['Ram'].str.replace('GB', '').astype('int32')
```

```python
data['Weight'] = data['Weight'].str.replace('kg','').astype('float64')
```

```python
data.head(2)
```

```python
data = data.drop(columns=['Inches'])
```

```python
data.corr(numeric_only=True)['Price_euros']
```

```python
data.head(2)
```

```python
data['Company'].value_counts()
```

```python
def add_company(company):
    if company=='Dell' or company=='Lenovo' or company=='HP' or company=='Asus' or company=='Acer' or company=='MSI' or company=='Toshiba' or company=='Apple' :
        return company
    else:
        return 'other'

data['Company'] =  data['Company'].apply(add_company)       
```

```python
data['Company'].value_counts()
```

```python
data['Product'].value_counts()
```

```python
data = data.drop(columns=['Product'])
```

```python
data['TypeName'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0 )
data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0 )
```

```python
data.head(2)
```

```python
data = data.drop(columns= ['ScreenResolution'])
```

```python
data['Cpu'].value_counts()
```

```python
data['Cpu_name'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
data['Cpu_name'].value_counts()
```

```python
def cpu_name(cpu):
    if  cpu == 'Intel Core i7' or cpu == 'Intel Core i5' or cpu == 'Intel Core i3' :
        return cpu
    else:
        if cpu.split()[0]=='AMD':
            return 'AMD'
        else :
            return 'other'

data['Cpu_name']= data['Cpu_name'].apply(cpu_name)            
```

```python
data['Cpu_name'].value_counts()
```

```python
data = data.drop(columns=['Cpu'])
```

```python
data.head(2)
```

```python
data['Gpu'].value_counts()
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x: ' ' .join(x.split()[0:1]))
```

```python
data['gpu_name'].value_counts()
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
data = data.drop(columns=['Gpu'])
```

```python
data['OpSys'].value_counts()
```

```python
def add_os(os):
    if os== 'Windows 10' or  os== 'Windows 7' or os== 'Windows 10 S'  :
        return 'Windows'
    elif os=='macOS' or os=='Mac OS X':
        return 'mac'
    elif os=='Linux' :
        return os
    else :
        return 'other'
        
data['OpSys'] = data['OpSys'].apply(add_os)    
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python

x= data.drop(columns=['Price_euros'])
y = data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y, test_size=0.25)
```

```python
def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_test,y_test)
    print(str(model)+  '->'+ str(acc))
    
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
ls= Lasso()
model_acc(ls)

from sklearn.tree import DecisionTreeRegressor
dt= DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor()
model_acc(rf)

```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':[10,50,100],
              'criterion':['squarred_error', 'absolute_error', 'poisson']}

grid_obj = GridSearchCV(estimator = rf, param_grid= parameters)
grid_fit = grid_obj.fit(x_train,y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
